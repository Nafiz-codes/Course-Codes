```python
def l(k,x):
  n = len(x)
  assert (k<(len(x)))
  x_k = x[k]
  x_copy = np.delete(x,k)
  denominator = np.prod(x_copy - x_k)
  coeff = []

  for i in range(n):
    coeff.append(sum([np.prod(x) for x in combinations(x_copy,i)])*(-1)**(i)/denominator)
  coeff.reverse()
  return Polynomial(coeff)
```

```python
def t(x):
  return (x-0)/(1-0)
```

```python
def h(k,x):
  temp = t(x)
  if x==0:
    coeff = np.array([1, -3, 2])
    p = Polynomial(coeff)
    return p(temp)
  elif x==1:
    coeff = np.array([0, -3, 2])
    p = Polynomial(coeff)
    return p(temp)
```

```python
def h_hat(k,x):
  temp = t(x)
  if x==0:
    coeff = np.array([0, 1, -2, 1])
    p = Polynomial(coeff)
    return p(temp)
  elif x==1:
    coeff = np.array([0, 0, -1, 1])
    p = Polynomial(coeff)
    return p(temp)
```

```python
def hermit(x,y,y_prime):
  assert len(x) == len(y)
  assert len(y) == len(y_prime)

  f = Polynomial([0])
  for i in range(len(x)):
    f+= ((y[i]*h(i,x[i]))+(y_prime[i]*h_hat(i,x[i])))

  return f
```
